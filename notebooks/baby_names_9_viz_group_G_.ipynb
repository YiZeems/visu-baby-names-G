# Final brainstorming selection — 9 visualizations for *Baby Names*

This notebook presents the final selection after the brainstorming phase: **3 graphs per problem**, for a total of **9 visualizations**.

- **Problem 1: temporal evolution of baby names** → **1.6, 1.8, 1.2**
- **Problem 2: regional effects** → **2.1, 2.9, 2.4**

For each selected visualization, we explain what it shows, how it works, why it was kept, and what its limitations are.


# Overall brainstorming logic

At first, we generated many different ideas for each problem. Some were standard, such as maps, curves, rankings, or scatterplots. Others were more creative, such as the leaf timeline, the name genealogy, the regional flower, or icon-based gender sketches. After discussion, we decided that the final selection should not rely only on the most original ideas, because they can sometimes be less precise. It should also not rely only on standard graphs, because the goal of the design phase is to explore alternatives.

We therefore selected **three graphs for each problem** with a complementary logic. For each question, we kept at least one more creative or narrative design, one clear and readable design, and one more analytical design. This gives a balanced selection: the ideas can communicate visually, support comparison, and answer the project questions.

# Problem 1 — Temporal evolution

For this problem, we want to understand how names change between 1900 and 2020: appearance, popularity peak, decline, comeback, or replacement by other names. The selected visualizations are **1.6**, **1.8**, and **1.2**.


## 1.6 — Leaf timeline / visual popularity frieze

Shows the life cycle of a name along a timeline.

**What it shows:** Visual marks such as leaves or symbols are placed on the time axis to represent periods of stronger popularity.

**How it shows it:** It is original, memorable, and good for communicating the idea of growth, decline, and comeback.

**Strengths:** Original, memorable, and good for communicating growth, decline, and comeback.

**Limitations:** Less precise than a standard statistical chart.



**Why we kept it:** We kept this idea because it gives an intuitive view of a name's life cycle: appearance, growth, peak, decline, or comeback. Its strength is originality and storytelling.


## 1.8 — Name genealogy / tree

Shows names as generations or families evolving over time.

**What it shows:** Branches connect names or name families across periods, suggesting historical succession.

**How it shows it:** It supports a narrative interpretation of name evolution and style replacement.

**Strengths:** Supports a narrative interpretation of name evolution and style replacement.

**Limitations:** Links between names may become subjective if grouping criteria are not clearly defined.



**Why we kept it:** We kept it because it supports a historical and narrative interpretation of name evolution, showing how some names dominate a period before being replaced by others.


## 1.2 — Bump chart of ranks

Shows how the rank of each name changes over time.

**What it shows:** Each line follows one name across years or decades. Moving upward means the name becomes more popular in the ranking.

**How it shows it:** It clearly shows which names rise, fall, enter the top ranks, or disappear.

**Strengths:** Very clear for seeing which names rise, fall, or stay in the top ranks.

**Limitations:** Shows relative rank only, not the actual difference in number of births.



**Why we kept it:** We added the bump chart because it complements the two more narrative temporal ideas. It clearly shows rank changes over time and is useful for comparing names directly.


## Justification for Problem 1

We kept **1.6** and **1.8** because they tell the story of name evolution in a more original way than standard curves. However, they are less precise for comparison, so we added **1.2**, the bump chart, which clearly shows rank changes over time. Together, the three ideas combine narrative, originality, and analysis.


# Problem 2 — Regional effects

For this problem, we want to understand whether certain names are more popular in specific regions or departments. The selected visualizations are **2.1**, **2.9**, and **2.4**.


## 2.1 — Department choropleth map

Shows how popular a name is in each French department.

**What it shows:** Each department is colored according to the name’s birth share or normalized popularity.

**How it shows it:** It is the most intuitive representation for a geographic question.

**Strengths:** The most intuitive representation for a geographic question.

**Limitations:** Raw counts can be misleading because highly populated departments would dominate; normalized values are required.



**Why we kept it:** We kept it because it is the most natural visualization for a regional question.


## 2.9 — Regional flower

Summarizes the regional profile of a name.

**What it shows:** Each petal represents a region, and the petal length encodes the name’s popularity in that region.

**How it shows it:** It is original, compact, and easy to compare across names.

**Strengths:** Original, compact, and easy to compare regional profiles across names.

**Limitations:** Loses department-level precision and exact geography.



**Why we kept it:** We kept it because it is more original than a map and makes regional profiles easy to compare.


## 2.4 — Department ranking

Shows which departments are most associated with a given name.

**What it shows:** Departments are ordered by the popularity or percentage of births for the selected name.

**How it shows it:** It is very readable and gives precise top regions.

**Strengths:** Very readable and gives precise top departments or regions.

**Limitations:** Loses the geographic shape of France.



**Why we kept it:** We added the department ranking because it complements the map. A map shows where the pattern occurs, but small departments may be hard to read; the ranking lists the strongest departments directly.


## Justification for Problem 2

We kept **2.1** because a map is essential for a geographic question. We also kept **2.9** because the regional flower gives a more original and synthetic profile view. Finally, we added **2.4** because the ranking makes the comparison more readable and identifies the strongest departments directly. Together, the three designs provide location, regional profile, and precise ranking.


# Problem 3 — Gender effects

For this problem, we want to show whether names are mostly given to girls, boys, or both, and how that balance can change over time. The selected visualizations are **3.3**, **3.8**, and **3.6**.


## 3.3 — Baby ratio sketch

Shows the girls/boys ratio for a few names in a very intuitive way.

**What it shows:** A baby icon is colored according to the proportion of girls and boys.

**How it shows it:** It is immediately understandable for a non-expert audience.

**Strengths:** Immediately understandable for a non-expert audience.

**Limitations:** Not very precise and works mainly for a small number of examples.



**Why we kept it:** We kept it because it is very intuitive and easy to understand. It works well for explaining the concept of gender ratio to a non-expert audience.


## 3.8 — Gender trajectory

Shows whether a name becomes more feminine, more masculine, or remains mixed.

**What it shows:** The vertical axis represents years, and the horizontal axis represents the girls/boys ratio.

**How it shows it:** It is clear for following one unisex name over time.

**Strengths:** Clear for following one unisex name over time.

**Limitations:** Works mainly name by name, not for many names at once.



**Why we kept it:** We kept the gender trajectory because it clearly shows whether a unisex name becomes more feminine, more masculine, or stays mixed over time. It complements the intuitive baby sketch with a precise temporal view.


## 3.6 — Popularity × mixedness scatterplot

Places names according to both total popularity and gender mixedness.

**What it shows:** The x-axis shows total popularity, and the y-axis shows how balanced the name is between girls and boys.

**How it shows it:** It identifies names that are both popular and truly unisex.

**Strengths:** Identifies names that are both popular and truly unisex.

**Limitations:** The mixedness measure must be explained, and overlapping points may need filtering.



**Why we kept it:** We added the popularity × mixedness scatterplot because it provides a more global and analytical view. It helps distinguish truly mixed names from names that are simply rare.


## Justification for Problem 3

We kept **3.3** because it makes gender proportions very accessible and easy to understand. We added **3.8**, the gender trajectory, to show how a unisex name drifts between girls and boys over the years. Finally, **3.6** provides a global analytical view of popularity and mixedness across all names. This combination balances simple communication, temporal analysis, and dataset-wide comparison.
