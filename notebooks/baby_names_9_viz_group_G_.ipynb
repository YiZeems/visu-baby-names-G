# Final brainstorming selection — 9 visualizations for *Baby Names*

This notebook presents the final selection after the brainstorming phase: **3 graphs per problem**, for a total of **9 visualizations**.

- **Problem 1: temporal evolution of baby names** → **1.6, 1.8, 1.2**
- **Problem 2: regional effects** → **2.1, 2.8, 2.3**
- **Problem 3: gender effects** → **3.3, 3.7, 3.6**

For each selected visualization, we explain what it shows, how it works, why it was kept, and what its limitations are.


# Overall brainstorming logic

At first, we generated many different ideas for each problem. Some were standard, such as maps, curves, rankings, or scatterplots. Others were more creative, such as the leaf timeline, the name genealogy, the regional flower, or icon-based gender sketches. After discussion, we decided that the final selection should not rely only on the most original ideas, because they can sometimes be less precise. It should also not rely only on standard graphs, because the goal of the design phase is to explore alternatives.

We therefore selected **three graphs for each problem** with a complementary logic. For each question, we kept at least one more creative or narrative design, one clear and readable design, and one more analytical design. This gives a balanced selection: the ideas can communicate visually, support comparison, and answer the project questions.

# Problem 1 — Temporal evolution

For this problem, we want to understand how names change between 1900 and 2020: appearance, popularity peak, decline, comeback, or replacement by other names. The selected visualizations are **1.6**, **1.8**, and **1.2**.


## 1.6 — Leaf timeline / visual popularity frieze



This visualization shows the evolution of a name or a group of names along a timeline. Visual marks such as leaves represent the periods when the name becomes more popular. We kept this idea because it gives an intuitive view of a name’s life cycle: appearance, growth, peak, decline, or comeback. Its strength is originality and storytelling. Its limitation is that it is less precise than a standard line chart.


## 1.8 — Name genealogy / tree



This visualization represents names as a kind of tree or genealogy. The idea is to show successive generations of names and how some names dominate a period before being replaced by others. We kept it because it supports a historical and narrative interpretation of name evolution. Its limitation is that the links between names must be clearly justified, otherwise the structure can become subjective.


## 1.2 — Bump chart of ranks



We added the bump chart because it complements the two more narrative temporal ideas. It shows the rank of names across years or decades. A line going up means that a name gains popularity in the ranking, while a line going down means that it loses places. This is useful for comparing names directly. Its limitation is that it shows rank, not the actual difference in birth counts.


## Justification for Problem 1

We kept **1.6** and **1.8** because they tell the story of name evolution in a more original way than standard curves. However, they are less precise for comparison, so we added **1.2**, the bump chart, which clearly shows rank changes over time. Together, the three ideas combine narrative, originality, and analysis.


# Problem 2 — Regional effects

For this problem, we want to understand whether certain names are more popular in specific regions or departments. The selected visualizations are **2.1**, **2.8**, and **2.3**.


## 2.1 — Department choropleth map



The choropleth map shows the popularity of a name in each department. Each department is colored according to the observed value, for example the percentage of births with that name. We kept it because it is the most natural visualization for a regional question. Its limitation is that raw counts can be misleading, so a normalized measure should be used.


## 2.8 — Regional flower



The regional flower summarizes the regional profile of a name. Each petal represents a region, and the petal length indicates the name’s popularity in that region. We kept it because it is more original than a map and makes regional profiles easy to compare. Its limitation is that it loses department-level precision and exact geography.


## 2.3 — Department ranking



We added the department ranking because it complements the map. A map shows where the pattern occurs, but small departments may be hard to read. The ranking directly lists the departments where the name is most represented. Its limitation is that it loses the geographic shape, so it works best alongside a map.


## Justification for Problem 2

We kept **2.1** because a map is essential for a geographic question. We also kept **2.8** because the regional flower gives a more original and synthetic profile view. Finally, we added **2.3** because the ranking makes the comparison more readable and identifies the strongest departments directly. Together, the three designs provide location, regional profile, and precise ranking.


# Problem 3 — Gender effects

For this problem, we want to show whether names are mostly given to girls, boys, or both. The selected visualizations are **3.3**, **3.7**, and **3.6**.


## 3.3 — Baby ratio sketch



The baby ratio sketch represents a name using a colored baby icon, where the colors show the girls/boys proportion. We kept it because it is very intuitive and easy to understand. It works well for explaining the concept of gender ratio. Its limitation is that it is not very precise and mainly works for a few examples.


## 3.7 — Gender mini-cards



Gender mini-cards compare several names or periods using small visual summaries. Each card shows the girls/boys ratio for one case. We kept this idea because it is simple, readable, and more comparative than a single icon. Its limitation is that it can take a lot of space if too many names are shown.


## 3.6 — Popularity × mixedness scatterplot



We added the popularity × mixedness scatterplot because it provides a more global and analytical view. Each point represents a name. The x-axis shows total popularity, and the y-axis shows the degree of gender mixedness. This helps distinguish truly mixed names from names that are simply rare. Its limitation is that the mixedness index must be clearly explained.


## Justification for Problem 3

We kept **3.3** and **3.7** because they make gender proportions very accessible and easy to understand. However, they mostly work on selected examples, so we added **3.6**, which gives a more global view of all names. This combination balances simple communication with deeper analysis.


# General conclusion

The final selection contains **9 visualizations**, with **3 proposals for each problem**. The strength of this selection is that it does not rely on only one type of chart. Each problem is approached through several angles: narrative, comparison, and analysis.

For the next stage, the strongest candidates for implementation are likely the **bump chart** for temporal evolution, the **choropleth map with a department ranking** for regional effects, and the **popularity × mixedness scatterplot** for gender effects. The other designs remain useful for communicating the brainstorming process and showing the diversity of design alternatives.
